```python
import matplotlib.pyplot as plt

# Dataset
people = ['shiva', 'harsha', 'chinna', 'mani']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]



# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(height, weight, color='blue', marker='o')

# Adding Titles and Labels
plt.title('Height vs Weight of Individuals')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

# Display Plot
plt.grid(True)
plt.show()




# Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(people, weight, color='green')

# Adding Titles and Labels
plt.title('Weight of Individuals')
plt.xlabel('Person')
plt.ylabel('Weight (kg)')

# Display Plot
plt.show()




# Histogram
plt.figure(figsize=(10, 6))
plt.hist(age, bins=4, color='orange', edgecolor='black')

# Adding Titles and Labels
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Display Plot
plt.show()
```
